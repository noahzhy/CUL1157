```python
import re
import requests
from datetime import datetime, timedelta, timezone

## static methods

def clean_strings(string:str) -> str:
    # remove the html tag in strings
    dr = re.compile(r'<[^>]+>',re.S)
    return dr.sub('', string)


def days_ago(t:int=30) -> str:
    """
    Only that one day
    """
    # free api request only support get news one month recently
    t = 30 if t > 30 else t
    _date = (datetime.now(tz=timezone.utc) - timedelta(t)).strftime("%Y-%m-%d")
    print("from", _date, "to", _date)
    return "from={}&to={}&".format(_date, _date)
```

```python
keywords = "google"
from2to = days_ago(30)
page_size = 100
```

```python
url = (
    'http://newsapi.org/v2/everything?'
    'q={}&'.format(keywords)+
    from2to+
    'pageSize={}&'.format(page_size)+
    'language=en&'
    'sortBy=popularity&'
    'apiKey=528b738755984880b5bd6198f4bea83e'
)

response = requests.get(url)
articles = response.json()["articles"]
print("articles", len(articles))

for i in articles:
    title = clean_strings(i["title"])
    description = clean_strings(i["description"])
    print(title, description)
    print(i["publishedAt"])
    print('='*50)
```
